To analyses sales data form each month, we categorise all sales by data and get the sum

# Clean data and remove nans

```python
dff = dff.dropna(how='all')
```

# Find 'Or' and delete, duplicating colum headers

```python
dff = dff[dff['Order Date'].str[0:2]!='Or']
dff
```

# Add month column and sales

```python
dff['Month'] = dff['Order Date'].str[0:2]
dff['Month'] = dff['Month'].astype('int32')
dff
```

# to create sales column by first converting quntity and price columns

```python
dff['Quantity Ordered'] =pd.to_numeric(dff['Quantity Ordered'])
dff['Price Each']= pd.to_numeric(dff['Price Each'])
```

```python
dff['Sales'] = dff['Quantity Ordered'] * dff['Price Each']
dff
```

# Question : What is the best month for sales and how much was earned

```python
saleforecast = dff.groupby('Month').sum()
saleforecast
```

```python
saleforecast.to_csv('sales.csv')
```

# Question: Which city had the highest sales

#add a city column

```python
dff
```

```python
dff['City'] = dff['Purchase Address'].apply(lambda x: x.split(',')[1])
```

```python
dff
```

#Add state for cities

```python
def get_state(address):
    return address.split(',')[2].split(' ')[1]
dff['State'] = dff['Purchase Address'].apply(lambda x: get_state(x)) 
```

```python
dff
```

# Which city had the highest sells

```python
city_sales = dff.groupby('City').sum()
```

```python
city_sales.to_csv('city_sales.csv')
```

# What products are often sold together

```python
duplicated_orders = dff[dff['Order ID'].duplicated(keep=False)]
duplicated_orders
```

```python
duplicated_orders['Grouped'] = duplicated_orders.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
duplicated_orders = duplicated_orders[['Order ID', 'Grouped']].drop_duplicates()
```

```python
duplicated_orders.head(10)
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in duplicated_orders['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
count.most_common(10)
```
